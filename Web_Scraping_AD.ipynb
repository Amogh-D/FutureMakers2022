```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.fsis.usda.gov/recalls-alerts/smithfield-packaged-meats-corp--recalls-a-ready-eat-bacon-topping-products-due'

req = requests.get(url)
soup = BeautifulSoup(req.text,'html.parser')

for s in soup.find_all('h1',class_='recall-header__title'):
  print(s.text)

for s in soup.find_all('a',class_ = 'tag tag--reason'):
  print(s.text)

for s in soup.find_all('span',class_='tag tag--active'):
  if "Active" not in str(s.text):
    print(s.text)
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.fsis.usda.gov/recalls'

req = requests.get(url)
soup = BeautifulSoup(req.text, 'html.parser')

links = []
for s in soup.find_all('a',href=True):
  if 'recalls-alerts' in str(s):
    links.append('https://www.fsis.usda.gov'+s['href'])

links = list(set(links))
links
```

```python
report_links = []
recall_reason = []
report_title = []
report_id = []

for i in links:

  req = requests.get(i)
  soup = BeautifulSoup(req.text, 'html.parser')


  for s in soup.find_all('h1',class_='recall-header__title'):
    report_title.append(s.text)

  for s in soup.find('a',class_ = 'tag tag--reason'):
    recall_reason.append(s)

  for s in soup.find_all('span',class_='tag tag--active'):
    if "Active" not in str(s.text):
      report_id.append(s.text)

  report_links.append(i)
```

```python
import pandas as pd

d = {'Report-ID':report_id, 'Report-Title': report_title, 'Recall-Reason':recall_reason, 'Report-URL':report_links}
df = pd.DataFrame(data=d)
df.head(50)

#df.to_excel('fsis_data.xlsx')
```
